## Creating the oscilloscope object

```python
from utils import *
from osctck import *

TDS2024 = Osctck('USB0::1689::874::C034414::0::INSTR')
```

## Oscilloscope configuration and data acquisition

```python
TDS2024.config(channels = (1,),
                 triggerSource = 'EXT',
                 triggerLevel = -0.4,
                 triggerSlope = 'FALL',
                 triggerMode = 'NORM',
                 triggerCoup = 'AC',
                 acquisition = 128,
                 vAutoScale = True)

meas = TDS2024()
```

## Plot of results

```python
import matplotlib.pyplot as plt

plt.plot(meas[0], meas[1])
plt.ylabel('Amplitud  CH1 [V]')
plt.xlabel('Tiempo [s]')
plt.show()
```

## Saving results

```python
txtName = getTxtName()
np.savetxt("Mediciones/Campo/" + txtName + ".txt", np.column_stack((meas[0], meas[1])),fmt='%.12f')
```
