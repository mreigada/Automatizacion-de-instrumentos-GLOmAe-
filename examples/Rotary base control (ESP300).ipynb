## Creating the rotating base object

```python
from rotmcESP import *

ESP300 = RotmcESP('ASRL/dev/ttyUSB0::INSTR')
```

## Motor and rotation controller configuration

```python
ESP300.config(axis = 2,
              vel = 5,
              direction = '+',
              setOrigin = True)

currPostion = ESP300(reference = 'REL', rotAngle = 360)

print(currPostion)
```
